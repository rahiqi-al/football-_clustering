```python
df_player['metrique_performance']=df_player['overall_rating']*0.7+df_player['potential']*0.3
```

```python
df_player.head()
```

```python
df_player.metrique_performance.isnull().sum()
```

```python
def plus_performants(group):
    sorted_group = group.sort_values(by='metrique_performance', ascending=False)
    # display(sorted_group)
    return sorted_group.head(1)


df_player_plus_performants = df_player.groupby('season').apply(plus_performants).reset_index(drop=True)
```

```python
df_player_plus_performants.head(15)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(df_player_plus_performants.head(6)['season'].astype(str), df_player_plus_performants.head(6)['metrique_performance'], color='skyblue')

# Adding player names above the bars
for _, row in df_player_plus_performants.head(6).iterrows():
    plt.text(row['season'], row['metrique_performance'] + 1, row['player_name'], ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Season')
plt.ylabel('Métrique Performance')
plt.title('Best Player and Their Score per Season')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# we need to calculate the std to see if the performance stay steady or change a lot (the smaller the better)

df_player_constant_profermonce=df_player.groupby('player_name')['metrique_performance'].std().reset_index()
df_player_constant_profermonce.columns=['player_name','std_of_performance']
display(df_player_constant_profermonce.sort_values(by='std_of_performance'))
df_player_constant_profermonce=df_player_constant_profermonce.sort_values(by='std_of_performance')
# the result make sense cause its the most consistants performance doesnt mean his the best
display(df_player_constant_profermonce[df_player_constant_profermonce['player_name']=='Lionel Messi'])
# above to make sure
#display(df_player_constant_profermonce)
```

```python
player_canstant_performance=list(df_player_constant_profermonce.player_name.values)
player_canstant_performance
```

```python
df_player['season'] = df_player['season'].apply(lambda x : int(x.split('/')[0]))
df_player.head()
#display(df_player[df_player['player_name']=='Lionel Messi'])
```

```python
plt.figure(figsize=(10, 6))
for player in player_canstant_performance[:4]:
    player_data = df_player[df_player["player_name"] == player]
    plt.plot(
        player_data["season"],
        player_data["metrique_performance"],
        marker="o",
        label=f"Player {player}",
    )

# Add titles and labels
plt.title("player performance trends over seasons", fontsize=14)
plt.xlabel("season", fontsize=12)
plt.ylabel("performance score", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(title="players", fontsize=10)
plt.tight_layout()
plt.show()
```

```python
def compare_team(group):
    return group[['buildUpPlaySpeed','chanceCreationPassing','defencePressure']].median()

df_compare_team=df_team.groupby('team_long_name').apply(compare_team).reset_index()
display(df_compare_team)
```

```python
df_subset = df_compare_team.head(7)

# Positions for each team
x = np.arange(len(df_subset['team_long_name']))  # Positions for each team

# Bar width
bar_width = 0.25

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x - bar_width, df_subset['buildUpPlaySpeed'], width=bar_width, label='Build-Up Speed')
ax.bar(x, df_subset['chanceCreationPassing'], width=bar_width, label='Passing Accuracy')
ax.bar(x + bar_width, df_subset['defencePressure'], width=bar_width, label='Defensive Pressure')

ax.set_xlabel('Teams')
ax.set_ylabel('Values')
ax.set_title('Comparison of Teams Based on Playing Styles')
ax.set_xticks(x)
ax.set_xticklabels(df_subset['team_long_name'], rotation=45)
ax.legend()

plt.tight_layout()
plt.show()
```
