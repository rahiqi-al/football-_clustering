# ------------------------player-------------------

```python
df_player
```

```python
df_player['metrique_performance']=df_player['overall_rating']*0.7+df_player['potential']*0.3
```

```python
df_player.head()
```

```python
df_player.metrique_performance.isnull().sum()
```

```python
def plus_performants(group):
    sorted_group = group.sort_values(by='metrique_performance', ascending=False)
    # display(sorted_group)
    return sorted_group.head(1)


df_player_plus_performants = df_player.groupby('season').apply(plus_performants).reset_index(drop=True)
```

```python
df_player_plus_performants.head(15)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.bar(df_player_plus_performants['season'].astype(str), df_player_plus_performants['metrique_performance'], color='skyblue')

# Adding player names above the bars
for _, row in df_player_plus_performants.iterrows():
    plt.text(row['season'], row['metrique_performance'] + 1, row['player_name'], ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Season')
plt.ylabel('Métrique Performance')
plt.title('Best Player and Their Score per Season')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# we need to calculate the std to see if the performance stay steady or change a lot (the smaller the better)

df_player_constant_profermonce=df_player.groupby('player_name')['metrique_performance'].std().reset_index()
df_player_constant_profermonce.columns=['player_name','std_of_performance']
display(df_player_constant_profermonce.sort_values(by='std_of_performance'))
# the result make sense cause its the most consistants performance doesnt mean his the best
display(df_player_constant_profermonce[df_player_constant_profermonce['player_name']=='Lionel Messi'])
# above to make sure
```

##### _________ Analyse des Tendances _____________

```python
df_player['season'] = df_player['season'].apply(lambda x : int(x.split('/')[0]))
df_player.head()
#display(df_player[df_player['player_name']=='Lionel Messi'])
```

```python
plt.figure(figsize=(10, 6))
for player_id in df_player["player_name"].unique()[:4]:
    player_data = df_player[df_player["player_name"] == player_id]
    plt.plot(
        player_data["season"],
        player_data["metrique_performance"],
        marker="o",
        label=f"Player {player_id}",
    )

# Add titles and labels
plt.title("Player Performance Trends Over Seasons", fontsize=14)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Performance Score", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(title="Players", fontsize=10)
plt.tight_layout()
plt.show()
```
